Todo el sistema de scraping en una celda

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from urllib.parse import urljoin # url general y adapta a una dinamica
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager # descarga y gestiona automaticamente la version corercta del driver
from selenium.webdriver.common.by import By #sirve para localizar elementos en la pagina(por CSS, ID, etc) algoritmo de busqueda
from selenium.webdriver.support.ui import WebDriverWait #permite hacer esperas explicitas hasta que algo este presente / completado
from selenium.webdriver.support import expected_conditions as EC # Define las condiciones que se esperan(elementos, visibilidad, etc)

options = Options() #c onfiguracion ordenador modo headless

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=options)

titulos, precios, stocks = [], [], []

url = "https://books.toscrape.com/"

try:
    while True:

        driver.get(url) #cargar la pagina en el navegador
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "article.product_pod"))) # espera explicita a que todoso los valores "article.product_pod" esten cargados

        soup = bs(driver.page_source, "html.parser") # Guarda y procesa el html en soup con selenium y beautifulsoup directamente

        articles = soup.select("article.product_pod") # separar los componentes "article.prduct_pod"

        for libro in contenido_soup.find_all("article", attrs={"class":"product_prod"}):

            titulo = libro.h3.a["title"].strip()
            titulos.append(titulo)
            
            precio = libro.find("p", class_="price_color").get_text(strip=True)
            precios.append(precio)
            
            stocks = libro.find("p", class_="instock availability").get_text(strip=True)
            stocks.append(stock)
    #buscar el enlace a la siguiente pagina(si existe)
        next_link = soup.select_one("li.next > a")
        if next_link:
            url = urljoin(url, next_link.get("href"))
        else:
            break
    libros_df = pd.DataFrame({"Titulo":titulos,"Precio":precios,"Stock":stocks})

    libros_df.to_csv("libros_prueba.csv", index=False, encoding="utf-8")
finally:
    driver.quit()
```
